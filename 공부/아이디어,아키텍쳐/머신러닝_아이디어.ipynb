```python
"""
CNN : 이미지나 공간 패턴을 자동을 특징 추출하는 모델

ReLU: 신경망이 복잡한 패턴을 배우게 만드는 스위치

회귀: 숫자를 예측 (고장까지 남은 시간, 불량률, 수명)

분류(로지스틱): 정상 / 이상 / 불량 판단

인공신경망: 규칙을 사람이 안 짜고, 데이터로 학습


****************************************
스마트 팩토리 현장 (3가지 만 한다.)
1. 이상인가? (Classification)
지금 설비 상태 정상 / 이상?

2. 언제 고장나나? (Regression)
고장까지 남은 시간 예측

3. 패턴이 달라졌나? (Anomaly Detection)
평소랑 다른 움직임인가? 



********************************************
CNN 은 사진 분류용 만이 아니다.
--> 패턴이 공간적으로 연결된 데이터면 다 CNN 이다
--> 현장에서 CNN 쓰는 실제 예시
1. 불량검사 (진짜 많이 씀)
-용접면 사진, 표면 균열, 스크래치
결과 : 정상/ 미세불량 / 심각불량   --> 이건 CNN/Softmax 분류

2. 센서 데이터도 CNN 씀
예시 : 진동 센서(시간 x 주파수)
       온도 분포
       열화상
1D CNN / 2D CNN
구조 : 센서 데이터 -> 이미지처럼 변환 -> CNN


*************************************************
ReLU 는 왜 필요한데? 
ReLU 가 없으면? 신경망이 직선 계산기, 복잡한 패턴 못 배움
ReLU 가 있으면? 
--> 이 신호 중요하면 살리고, 아니면 꺼라 -> 비선형 관계 학습 가능

예: 진동이 5 이상일 때만 위험
    온도는 80 넘을 때 급격히 악화
--> 이런 꺾이는 지점을 ReLU가 처리함

****************************************************
회귀 VS 분류 

회귀 : 결과가 숫자
--> 고장까지 남은 시간(시간), 불량률 (%), 에너지 사용량
ex) 예측 고장까지 36시간

분류 : 결과가 상태
--> 정상/ 경고 / 위험 
kpi 에서 AI 판정: 위험 이런식으로 적히게

****************************************************
인공 신경망이 왜 필요한가
데이터 -> 학습 -> 스스로 규칙 생성
새로운 패턴에도 적응
임계치 자동 학습


*****************************************************
AI는 혼자 존재하지 않는다.
-> KPI + 대시보드 + 의사결정으로 연결된다

예시 프로젝트 1 (CNN)
- AI 역할 : 설비 이미지 불량 분류
- 모델 : CNN + ReLU + Softmax
결과 : 불량 확률 : 82%
대시보드: 최근 불량 이미지, 불량 유형별 비율, AI 신뢰도


예시 프로젝트 2 (회귀 + 신경망)
- AI 역할 : 고장까지 남은 시간 예측
- 모델 : MLP/LSTM
결과 : 예측 수명 42시간
대시보드 : 시간별 감소 그래프, 위험 구간 표시

예시 프로젝트 3 (이상치 탐지)
- AI 역할 : 평소와 다른 센서 패턴 감지
- 모델 : AutoEncoder / Isolation Forest
결과: 이상점수 0.87


****************************************************
CNN 은 설비 이미지나 센서 패턴에서 사람이 정의하기 어려운 특징을 자동 추출하고,
ReLU 같은 활성화 함수로 비선형 현상까지 학습해 불량 분류나 고장 예측에 활용됩니다






"""
```
